```python
import pandas as pd
import ranx
```

```python
hpo = pd.read_csv('genes_diabetes_hpo.csv')
disgenet = pd.read_csv('genes_diabetes_disgenet.csv')
```

```python
qrels_dict={}
run_dict= {}
d1 = dict(zip(disgenet.geneSymbol, disgenet.score))
d2 = dict(zip(hpo.SYMBOL, hpo.Score))
qrels_dict['q_1'] = d1
run_dict['q_1'] = d2
```

```python
qrels = ranx.Qrels(qrels_dict)
run = ranx.Run(run_dict)
```

```python
ranx.evaluate(qrels, run, ["map@5", "mrr"])
```
